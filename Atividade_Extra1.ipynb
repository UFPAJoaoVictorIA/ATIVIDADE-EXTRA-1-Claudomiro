## Exemplo 4.1

Uma memória RAM (MP) tem um espaço máximo de endereçamento de 2K. Cada célula pode armazenar 16 bits. Qual o valor total de bits que podem ser armazenados nesta memória e qual o tamanho de cada endereço?

```python
import math

# Exemplo 4.1
N_celulas = 2 * 1024  # 2K
bits_por_celula = 16

total_bits = N_celulas * bits_por_celula
tamanho_endereco = math.ceil(math.log2(N_celulas))

print(f"Total de bits: {total_bits} bits")
print(f"Tamanho de cada endereço: {tamanho_endereco} bits")
```

## Exemplo 4.4

Um computador possui um RDM com capacidade de armazenar 32 bits e um REM com capacidade de armazenar 24 bits. Sabendo-se que em cada acesso são lidas duas células da memória RAM (MP) e que o barramento de dados (BD) tem tamanho igual ao da palavra, pergunta-se:

a) Qual é a capacidade máxima de endereçamento do microcomputador em questão?

b) Qual é o total máximo de bits que podem ser armazenados na memória RAM (MP)?

## Exemplo 4.5

Um computador possui um BE (barramento de endereços) com capacidade de permitir a transferência de 33 bits de cada vez. Sabe-se que o BD (barramento de dados) permite a transferência de quatro palavras em cada acesso e que cada célula da memória RAM (MP) armazena um oitavo (1/8) de cada palavra. Considerando que a memória RAM (MP) pode armazenar um máximo de 64G bits, pergunta-se:

a) Qual é a quantidade máxima de células que podem ser armazenadas na memória RAM (MP)?

b) Qual é o tamanho do REM e do BD existentes neste processador?

c) Qual é o tamanho de cada célula e da palavra desta máquina?

```python
import math

# Exemplo 4.5
# Dados fornecidos:
total_bits_memoria = 64 * (2**30)  # 64G bits
be_bits = 33 # Tamanho do barramento de endereços (BE)
palavras_por_acesso_bd = 4 # Quantidade de palavras transferidas pelo BD por acesso
fracao_celula_por_palavra = 1/8 # Cada célula armazena 1/8 de uma palavra

# Deduções e cálculos:

# Primeiramente, vamos determinar o tamanho de uma célula e de uma palavra.
# Sabemos que o número de endereços é 2 elevado ao tamanho do BE.
n_enderecos = 2**be_bits

# Assumindo que cada endereço da memória principal aponta para uma única célula,
# podemos encontrar o tamanho da célula (tamanho_celula_bits) dividindo o total de bits da memória
# pelo número total de endereços.
tamanho_celula_bits = total_bits_memoria / n_enderecos

# Com o tamanho da célula, podemos encontrar o tamanho da palavra (tamanho_palavra_bits)
# usando a relação que a célula armazena 1/8 da palavra.
tamanho_palavra_bits = tamanho_celula_bits / fracao_celula_por_palavra

# a) Qual é a quantidade máxima de células que podem ser armazenadas na memória RAM (MP)?
# É o total de bits na memória dividido pelo tamanho de cada célula.
qtd_max_celulas = total_bits_memoria / tamanho_celula_bits

# b) Qual é o tamanho do REM e do BD existentes neste processador?
# O tamanho do REM (Registrador de Endereço de Memória) é igual ao tamanho do BE.
rem_bits = be_bits
# O tamanho do BD (Barramento de Dados) é o número de palavras transferidas por acesso multiplicado pelo tamanho da palavra.
bd_bits = palavras_por_acesso_bd * tamanho_palavra_bits

# c) Qual é o tamanho de cada célula e da palavra desta máquina?
# Já calculados acima: tamanho_celula_bits e tamanho_palavra_bits.

print(f"a) Quantidade máxima de células: {int(qtd_max_celulas)} células")
print(f"b) Tamanho do REM: {rem_bits} bits, Tamanho do BD: {int(bd_bits)} bits")
print(f"c) Tamanho de cada célula: {int(tamanho_celula_bits)} bits, Tamanho da palavra: {int(tamanho_palavra_bits)} bits")
```

## Questão 6 e 7

6. Qual é a função do registrador de endereços de memória (REM)? E do registrador de dados de memória (RDM)?
7. Desreva os barramentos que interligam processador e MP, indicando função e direção do fluxo de sinais de cada um.

### Resposta 6
- **REM (Registrador de Endereço de Memória):** Armazena temporariamente o endereço de memória que será acessado para leitura ou escrita.
- **RDM (Registrador de Dados de Memória):** Armazena temporariamente o dado que acabou de ser lido da memória ou o dado que será gravado nela.

### Resposta 7
- **Barramento de Endereços (BE):** Transporta o endereço gerado pelo processador até a memória. Fluxo: Unidirecional (Processador -> Memória).
- **Barramento de Dados (BD):** Transporta o conteúdo/dado entre processador e memória. Fluxo: Bidirecional.
- **Barramento de Controle (BC):** Transporta sinais de sincronização e comando (ex: se a operação é Leitura ou Escrita). Fluxo: Bidirecional (predominantemente Processador -> Memória).

## Questão 21

Considere um sistema constituído de um processador – memória cache – memória principal, no qual o tempo de acesso processador/memória cache é de 8 ns e o tempo de acesso memória cache/memória principal é de 70 ns. Observando-se um intervalo correspondente a 100 acessos consecutivos do processador e que a eficiência da memória cache é de 96%, calcule o tempo médio de acesso do sistema.

```python
# Questão 21
t_cache = 8 # ns
t_mp = 70 # ns
eficiencia = 0.96

# Tempo médio = (acerto * t_cache) + (erro * (t_cache + t_mp))
t_medio = (eficiencia * t_cache) + ((1 - eficiencia) * (t_cache + t_mp))

print(f"Tempo médio de acesso: {t_medio:.2f} ns")
```

## Questão 26

Projete uma memória ROM de forma semelhante a que está mostrada na Fig. 4.18, considerando que ela possua 8 células de 4 bits cada uma. Podem ser assumidos quaisquer valores para armazenar nas 8 células.

### Resposta 26
Para uma ROM de 8 células de 4 bits (8x4):
- **Endereçamento:** 3 linhas (2³ = 8 endereços).
- **Saída:** 4 linhas de dados.

**Tabela de Conteúdo Exemplo:**
| Endereço (Bin) | Conteúdo (Hex) |
|---|---|
| 000 | A |
| 001 | 5 |
| 010 | F |
| 011 | 0 |
| 100 | C |
| 101 | 3 |
| 110 | E |
| 111 | 9 |

```python
import math

class ROM:
    def __init__(self, num_cells, bits_per_cell, content=None):
        if not (num_cells > 0 and (num_cells & (num_cells - 1)) == 0): # Check if num_cells is a power of 2
            raise ValueError("Number of cells must be a power of 2.")
        self.num_cells = num_cells
        self.bits_per_cell = bits_per_cell
        self.address_lines = int(math.log2(num_cells))

        # Initialize ROM content (list of integers, each representing a cell's value)
        # Max value for a cell is 2**bits_per_cell - 1
        self.content = [0] * num_cells
        if content:
            if len(content) > num_cells:
                raise ValueError(f"Content provided ({len(content)} items) exceeds ROM capacity ({num_cells} cells).")
            for i, val in enumerate(content):
                if not (0 <= val < (2**bits_per_cell)):
                    raise ValueError(f"Value {val} at index {i} exceeds {bits_per_cell}-bit capacity.")
                self.content[i] = val

    def read(self, address):
        if not (0 <= address < self.num_cells):
            raise IndexError(f"Address {address} is out of bounds for ROM with {self.num_cells} cells.")
        return self.content[address]

    def __str__(self):
        s = f"ROM Characteristics:\n"
        s += f"  Number of cells: {self.num_cells}\n"
        s += f"  Bits per cell: {self.bits_per_cell}\n"
        s += f"  Address lines: {self.address_lines}\n"
        s += "\nROM Content:\n"
        for i, val in enumerate(self.content):
            s += f"  Address {i:0{self.address_lines}b} ({i:2d}): {val:0{math.ceil(self.bits_per_cell/4)}X} (Hex)\n" # Display as Hex
        return s
```

```python
# Questão 26: Projete uma memória ROM com 8 células de 4 bits cada.

# Definir o conteúdo da ROM (valores de 4 bits)
# Usando os valores da tabela exemplo: A, 5, F, 0, C, 3, E, 9
# Convertendo Hex para Decimal: A=10, 5=5, F=15, 0=0, C=12, 3=3, E=14, 9=9
rom_content_values = [10, 5, 15, 0, 12, 3, 14, 9]

# Criar uma instância da ROM
my_rom = ROM(num_cells=8, bits_per_cell=4, content=rom_content_values)

# Exibir as características e o conteúdo da ROM
print(my_rom)

# Demonstração de leitura de alguns endereços
print("\nDemonstração de leitura:")
print(f"Valor no endereço 000 (0 decimal): {my_rom.read(0):0{math.ceil(my_rom.bits_per_cell/4)}X} (Hex)")
print(f"Valor no endereço 010 (2 decimal): {my_rom.read(2):0{math.ceil(my_rom.bits_per_cell/4)}X} (Hex)")
print(f"Valor no endereço 111 (7 decimal): {my_rom.read(7):0{math.ceil(my_rom.bits_per_cell/4)}X} (Hex)")
```
